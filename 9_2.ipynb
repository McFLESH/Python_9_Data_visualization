```python
# Постройте линейный график

import matplotlib.pyplot as plt
def data_plot(x, y):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel('Ось X')
    ax.set_ylabel('Ось Y')
    ax.set_title('Линейный график')
    return ax
```
