```python
# Постройте два графика на одном окне


import matplotlib.pyplot as plt
def data_plots(x, y1, y2):
    plt.subplot(1, 2, 1)
    plot1, = plt.plot(x, y1, '-g')

    plt.subplot(1, 2, 2)
    plot2, = plt.plot(x, y2, '-.r')

    return plot1, plot2
```
